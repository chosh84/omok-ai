```python
FILE = '/content/drive/MyDrive/MLData/Omok-ai/dataset.zip'
FOLDER = '/content/drive/MyDrive/MLData/Omok-ai/dataset'

from zipfile import ZipFile
print('Unzipping zipfile...')
with ZipFile(FILE) as zip_data:
  zip_data.extractall(FOLDER)
```

```python
import numpy as np
from glob import glob
import os

base_path = os.path.join('/content/drive/MyDrive/MLData/Omok-ai/dataset/', '*.npz')

file_list = glob(base_path)

print(file_list)

for file_path in file_list:
  data = np.load(file_path)
  inputs = data['inputs']
  outputs = data['outputs']
  
  for input, output in zip(inputs, outputs):
    print(input)
    print(output)
    print('=======')
    
  print('=================')
  break
```
